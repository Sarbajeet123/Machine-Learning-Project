```python
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

for i in ast.literal_eval(obj):

ast.literal_eval(obj) takes obj, which is usually a string representation of a list of dictionaries,
for example:
"[{'id': 28, 'name': 'Action'}, {'id': 12, 'name': 'Adventure'}]"

After literal_eval, this string becomes a real Python list:
[{'id': 28, 'name': 'Action'}, {'id': 12, 'name': 'Adventure'}]
Then the for i in ... loop iterates over this list of dictionaries, assigning each dict to i.



```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

For cast(we want 3 Actors)

```python
import ast
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if(counter!=3):
           l.append(i['name'])
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

For crew(We want only Director Here)

```python
import ast
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
           l.append(i['name'])
           break
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

 y.append(ps.stem(i))

.stem() reduces the word to its root form.

The stemmed word is then appended to list y.

âœ… Explanation:

```python
new_df['tags']=new_df['tags'].apply(stem)
```

Vectorization....................................

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

we have converted all the words into vectors 

from sklearn.feature_extraction.text import CountVectorizer
âœ… Explanation:

You are importing the CountVectorizer class from the sklearn.feature_extraction.text module.

CountVectorizer is a tool to convert a collection of text documents into a matrix of token counts (basically, a bag-of-words model).

This means it takes text and turns it into numbers so machine learning models can work with it.


cv = CountVectorizer(max_features=5000, stop_words='english')
âœ… Explanation:

You are creating an instance of CountVectorizer named cv.

max_features=5000

This limits the vocabulary to the top 5,000 most frequent words across all documents.

So if the dataset has, say, 20,000 unique words, only the 5,000 most common ones are kept, which helps to reduce dimensionality.

stop_words='english'

This removes English stop words (like the, is, and, etc.) from the token list.

These words usually do not carry important meaning for many NLP tasks.

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

we are calculating the distance with the help of cosine function 1 means they are similar 0 means not similar 

Exactly â€” youâ€™ve got the idea.

If your vectors array represents, say, 4806 movies, where each movie is represented by a feature vector (e.g., TF-IDF of its plot, or an embedding), then:

similarity = cosine_similarity(vectors)
will produce a 4806 Ã— 4806 matrix, where:

similarity[i, j] measures how similar movie i is to movie j based on their vector representations.

So row 0 (i.e., similarity[0, :]) gives you the similarity of the first movie with all 4806 movies, including itself (which will always be 1, because the cosine similarity of a vector with itself is 1).

Thatâ€™s exactly how youâ€™d build a content-based recommender:
âœ… pick one movieâ€™s vector
âœ… measure cosine similarity to all other movies
âœ… recommend the movies with the highest similarity scores

```python
similarity[0]# first movie with other all movies
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movies):
    # 1. get the index of the movie you want to recommend for
    movie_index = new_df[new_df['title'] == movies].index[0]
    
    # 2. get all similarity scores for this movie
    distances = similarity[movie_index]
    
    # 3. pair every movie index with its similarity score and sort descending
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    # 4. print recommended movie indices
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
